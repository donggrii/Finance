```python
from numpy import dot           #  (cf. 1   dot  element-wise product)
from numpy.linalg import norm   # linear algebra norm(L2 norm default)
import numpy as np

#    
def cos_sim(A, B):
    return dot(A, B) / (norm(A) * norm(B))
```

```python
#  1 ~ 3  BoW 
doc1 = np.array([0, 1, 1, 1])
doc2 = np.array([1, 0, 1, 1])
doc3 = np.array([2, 0, 2, 2])

print(cos_sim(doc1, doc2))   #  1  2  
print(cos_sim(doc1, doc3))   #  1  3  
print(cos_sim(doc2, doc3))   #  2  3  
```
